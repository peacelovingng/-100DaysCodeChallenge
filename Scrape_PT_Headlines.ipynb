```python
import re
import requests
from bs4 import BeautifulSoup as bs
```

```python
p = requests.get('https://www.premiumtimesng.com/category/news/headlines?').content
t = requests.get('https://thenationonlineng.net').content
```

```python
pt_soup = bs(p, 'lxml')
tn_soup = bs(t, 'lxml')
```

```python
title1 = pt_soup.title.text
title1 = re.sub('Archives', ':', title1)
title1
```

```python
print(title1, '\n')
for i in pt_soup.find_all('h3', {'class':'a-story-title'}):
    print(i.text, '\n')
```

```python
title2 = tn_soup.title.text
title2
```

```python
print(title2, '\n')
for i in tn_soup.find_all('h3', {'class':'jeg_post_title'}):
    print(i.text)
```
